```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Student_Marks.csv')

feature = df['time_study']
target = df['Marks']

plt.scatter(feature, target)
plt.xlabel('Hours Studied per day')
plt.ylabel('Marks')
```

```python
def loss_function(w, b, points):
  total_error = 0
  for i in range (len(points)):
    x = points.iloc[i].time_study
    y = points.iloc[i].Marks
    y_hat = (w*x) + b
    total_error += (y - y_hat) ** 2
  total_error / float(len(points))
```

```python
def gradient_descent(w_now, b_now, points, L):
  w_gradient = 0
  b_gradient = 0

  n = len(points)

  for i in range(n):
    x = points.iloc[i].time_study
    y = points.iloc[i].Marks
    w_gradient += (-2/n) * x * (y - (w_now * x + b_now))
    b_gradient += (-2/n) * (y - (w_now * x + b_now))

  w = w_now - (w_gradient*L)
  b = b_now - (b_gradient*L)
  
  return w, b
```

```python
w = 0
b = 0
L = 0.0005
epochs = 1000

for i in range(epochs):
  if i%50 == 0:
    print(f"Epoch: {i}")
  w, b = gradient_descent(w, b, df, L)

print(w, b)

plt.scatter(feature, target)
plt.plot(feature, [w*x + b for x in feature], color = "black")
plt.show()
```
